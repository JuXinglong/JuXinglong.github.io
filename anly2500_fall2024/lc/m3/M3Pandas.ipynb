```python
city_pop = pd.DataFrame(
    [
        [808976, "San Francisco", "California"],
        [8363710, "New York", "New-York"],
        [413201, "Miami", "Florida"],
        [2242193, "Houston", "Texas"]
    ], index=[3,4,5,6], columns=["population", "city", "state"])


city_loc = pd.DataFrame(
    [
        ["CA", "San Francisco", 37.781334, -122.416728],
        ["NY", "New York", 40.705649, -74.008344],
        ["FL", "Miami", 25.791100, -80.320733],
        ["OH", "Cleveland", 41.473508, -81.739791],
        ["UT", "Salt Lake City", 40.755851, -111.896657]
    ], columns=["state", "city", "lat", "lng"])
```

```python
city_pop
```

```python
city_loc
```

```python
# concat stack vertically
df9=pd.concat([city_pop,city_loc])
df9
```

```python
pd.merge(left=city_loc,right=city_pop,left_on="city",right_on="city",how="inner") # horizontally inner, left, right, outer
```

```python
pd.merge(left=city_loc,right=city_pop,left_on="city",right_on="city",how="left")
```

```python
pd.merge(left=city_loc,right=city_pop,left_on="city",right_on="city",how="right")
```

```python
df10 = pd.merge(left=city_loc,right=city_pop,left_on="city",right_on="city",how="outer")
df10
```

```python
df11= df10.copy()
df11
```

```python
df12=df11.fillna(-1) # create a new dataframe
df12
```
